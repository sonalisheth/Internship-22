## 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
#### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
#### You need to find following details:
    A) Rank
    B) Name
    C) Artist
    D) Upload date
    E) Views


```python
# connect to webdriver
driver = webdriver.Chrome(r"C:/Users/DELL/Downloads/chromedriver_win32/chromedriver.exe")

# get webpage
driver.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.maximize_window()
```

```python
#creating empty lists
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []
```

```python
#finding all the rank tags
rank = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[1]')[0:30]
rank[0:3]
```

```python
#scraping the Rank
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]"):
    Rank.append(i.text)

#scraping the Name
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]"):
    Name.append(i.text.split('[')[0])

#scraping the Artist
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]"):
    Artist.append(i.text)

#scraping the upload date
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]"):
    Upload_date.append(i.text)

#scraping the number of views
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]"):
    Views.append(i.text)
    
```

```python
Rank=Rank[0:30]
Name=Name[0:30]
Artist=Artist[0:30]
```

```python
len(Rank),len(Name),len(Artist),len(Upload_date),len(Views)
```

```python
Yotube_Most_Views={'Rank':Rank,'Name':Name,'Artist':Artist,'Uploaded Date':Upload_date,'Views':Views}
Yotube_Most_Views=pd.DataFrame(data=Yotube_Most_Views)
Yotube_Most_Views
```

```python
driver.close()
```

```python
#scrap all names tag

names=[]
try:
    name=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
    for i in name:
        names.append(i.text)
except NoSuchElementException:
    names.append('N/A')
```

```python
# scrap all description tag

desc=[]
try:
    description=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/p")
    for i in description:
        desc.append(i.text)
except NoSuchElementException:
    desc.append('N/A')
```

```python
urls=[]
counts=[]
match_page=driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a')
for i in match_page:
    urls.append(i.get_attribute('href'))
```

```python
#Scrap all language tag

lang=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        listy=count.text.split('\n')
        if 'Languages' in listy:
            lang.append(listy[listy.index('Languages')+1])
        else:
            lang.append('N/A')
except NoSuchElementException:
    lang.append('N/A')
```

```python
counts=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        listy=count.text.split('\n')
        for j in listy:
            if j.split()[0]=='Contributors':
                counts.append(j.split()[1])
                break
        else:
            counts.append('N/A')
except NoSuchElementException:
    counts.append('N/A')
```

```python
len(names),len(desc),len(lang),len(counts)
```

```python
# create dataframe

dataframe=pd.DataFrame()
dataframe['Repository title']=names[0:24]
dataframe['Repository description']=desc[0:24]
dataframe['Language used']=lang[0:24]
dataframe['Contributors count']=counts[0:24]
dataframe
```
