### 1) Write a python program to display all the header tags from wikipedia.org.


```python
# send get request to the webpage server to get the source code of the page

page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
# content of the page 

soup = BeautifulSoup(page.content)
soup
```

```python
# scrapping all the header tags

header = []

for i in soup.find_all(['h1', 'h2','h3','h4','h5','h6']):
    header.append(i.text.replace('\n', ""))
    
print("Displaying all the Header Tags :\n")

res = "\n".join("{}".format(x) for x in zip(header))

print(res)
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
# send get request to the webpage server to get the source code of the page

imdb_1 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc') # top 50 movies
imdb_2 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt') # next top 50 vies
print(imdb_1)
print(imdb_2)
```

```python
# content of the page 

soup_1 = BeautifulSoup(imdb_1.content)
soup_2 = BeautifulSoup(imdb_2.content)
```

```python
# scraping all the names of the top 100 movie

name = []
for i in soup_1.find_all('h3', class_= 'lister-item-header'):
    name.append(i.text.replace("\n", "").split('.',)[1][:-6])
    
for i in soup_2.find_all('h3', class_= 'lister-item-header'):
    name.append(i.text.replace("\n", "").split('.',)[1][:-6])
    
print("Names of the top 100 movie are :\n")

res = "\n".join("{}".format(x) for x in zip(name))

print(res)
```

```python
# scraping all the ratings of the top 100 movie

rating = []

for i in soup_1.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n",""))
    
for i in soup_2.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n",""))
    
print("Rating of the top 100 movie are :\n")

res = "\n".join("{}".format(x) for x in zip(rating))

print(res)
```

```python
# scraping the year of release of the top 100 movie

year= []

for i in soup_1.find_all('span',class_='lister-item-year text-muted unbold'):
    year.append(i.text.replace("\n","").replace("(","").replace(")",""))
    
for i in soup_2.find_all('span',class_='lister-item-year text-muted unbold'):
    year.append(i.text.replace("\n","").replace("(","").replace(")",""))
    
print("Year of Release of the top 100 movie are :\n")

res = "\n".join("{}".format(x) for x in zip(year))

print(res)
```

```python
# lenghth of movie names, ratings and year of release

print(len(name),len(rating),len(year))
```

```python
# making dataframe 

import pandas as pd

df = pd.DataFrame({'Movie_Name': name, 'Rating': rating, 'Year_of_Release': year})
print("Below is the DataFrame of IMDB 'Top 100 Movies'\n")

df

```

### 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
# send get request to the webpage server to get the source code of the page

page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
# content of the page 

soup = BeautifulSoup(page.content)
soup
```

```python
# scraping all the Top 10 ODI teams in men’s cricket

teams = []

for i in soup.find_all('span',class_='u-hide-phablet'):
    teams.append(i.text)

teams = teams[:10]
print("Top 10 ODI teams in men’s cricket are :\n")

res = "\n".join("{}".format(x) for x in zip(teams))

print(res)
```

```python
# scraping all the 10 ODI teams in men’s cricket along with the records for matches

mat=soup.find('td',class_="rankings-block__banner--matches").text

matches=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)
    
Match=matches[::2]
Match.insert(0,mat)
Match=Match[:10]

print("Top 10 ODI teams in men’s cricket along with the records for matches are :")

res = "\n".join("{}\t{}".format(x, y) for x, y in zip(teams,Match))

print(res)
```

```python
# scraping all the 10 ODI teams in men’s cricket along with the records for matches and points

pt=soup.find('td',class_="rankings-block__banner--points").text

points=matches[1::2]
points.insert(0,pt)
points=points[:10]

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")

res = "\n".join("{}\t{}\t{}".format(x, y, z) for x, y, z in zip(teams,Match,points))

print(res)
```

```python
# scraping all the 10 ODI teams in men’s cricket along with the records for matches, points and rantings

ratings=[]

for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
     ratings.append(i.text.replace("\n","").replace(" ",""))
        
        
for i in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings.append(i.text.replace(" ","").replace("\n",""))
    
ratings=ratings[:10]

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")

res = "\n".join("{}\t{}\t{}\t{}".format(w,x, y, z) for w, x, y, z in zip(teams,Match,points,rating))

print(res)
```

```python
# making DataFrame

odiMenTeams=pd.DataFrame({"Team":teams,"Matches":Match,"Points":points,"Ratings":ratings})

print("Top 10 ODI teams in men’s cricket along with the records for matches and points are :")
odiMenTeams
```

### 9) Write a python program to scrape weather details for last 24 hours from Tutiempo.net :


```python
# send get request to the webpage server to get the source code of the page

page = requests.get("https://en.tutiempo.net/delhi.html?data=last-24-hours")
                    
# content of the page 

soup = BeautifulSoup(page.content)
```

#### i) Hour


```python
# scraping the hour data

hour=[]
for i in soup.find_all("div",class_="last24 thh"):
    hour.append(i.text.replace("Data New Delhi / Palam Climate.IcoTiempo {width:35px;}TodayHourWeather conditionWCTem.WindHum.Pressure","").replace("YesterdayHourWeather conditionWCTem.WindHum.Pressure","").split("hPa")[1:])

hour=hour[0]

hours=[]
for i in range(len(hour)):
    hours.append(hour[i][:5])
    
print("Displaying the hours data:")    
hours

```

#### ii) Temperature


```python
# scraping the temprature data

temp=[]
for i in soup.find_all('td',class_="t Temp"):
    temp.append(i.text)
    
print("Displaying the temprature data:")  

temp
```

#### iii) Wind


```python
# scraping the wind data

wind=[]
for i in soup.find_all('td',class_="wind"):
    wind.append(i.text.replace("En calma", "Calm"))
    
print("Displaying the Wind Data:")  

wind
```

#### iv) Weather condition


```python
# scraping the Weather condition data

weather=[]
for i in range(len(hour)):
    weather.append(hour[i][5:15].replace("°C","").replace("km/h","").replace("Calm","").replace("km/","").replace("k",""))
    
wc=[]
for i in weather:
    if "Mist" in i:
        i="Mist"
    elif "Drizzle" in i:
        i = "Drizzle"
    elif "Fog" in i:
        i = "Fog"
    wc.append(i)
    
print("Displaying the Weather condition Data:")  

wc
```

#### v) Humidity


```python
# scraping the humidity data

humidity=[]
for i in soup.find_all('td',class_="hr"):
    humidity.append(i.text)
    
print("Displaying the Humidity Data:")  

humidity
```

#### vi) Pressure

```python
# scraping the pressure data

pressure=[]
for i in soup.find_all('td',class_="prob"):
    pressure.append(i.text)
    
print("Displaying the Pressure Data:")  

pressure
```

```python
#Length of Hour, Temperature, Wind, Weather condition, Humidity, Pressure details for last 24 hours

print(len(hours),len(temp),len(wind),len(wc),len(humidity),len(pressure))
```
